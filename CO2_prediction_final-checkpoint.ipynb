```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import numpy as np
from scipy import stats

%matplotlib inline
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 150)
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (10, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

READING DATA

```python
df = pd.read_csv('/content/CO2 Emissions_Canada.csv')
df.head()
```

Make, Model and Vehical class columns doesn't have any use for prediction of co2 emission in vehicals so we drop this three columns. Fuel consumption comb(mpg) is drop because we alraedy have Fuel consumption comb in (liter per 100 Km), Fuel Consumption City (L/100 km) and Fuel Consumption Hwy (L/100 km) are also drop because average of Fuel consumption city and Fuel consumption Hwy is equal to Fuel Consuption combine (Fuel Consuption Comb)

```python
df.drop(columns=['Make', 'Model', 'Vehicle Class', 'Fuel Consumption Comb (mpg)', 'Fuel Consumption City (L/100 km)', 'Fuel Consumption Hwy (L/100 km)'], inplace=True)
df.head()
```

```python
df.isna().sum()
```

```python
df['Transmission'].unique()
```

```python
df['Fuel Type'].unique()
```

ENCODING NON-NUMERICAL COLUMNS

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
```

```python
tp = df[['Fuel Type', 'Transmission']].copy()
encoder.fit(tp)
encoded_cols = list(encoder.get_feature_names_out(['Fuel Type', 'Transmission']))
tp[encoded_cols] = encoder.transform(tp)
```

```python
df = pd.concat([df, tp], axis = 1)
```

```python
df
```

```python
df.drop(columns=['Fuel Type', 'Transmission'], inplace=True)
```

```python
df
```

***CHECKING CORRELATION***

```python

plt.figure(figsize=(30, 35))
corr = df.corr()
mask = np.zeros_like(corr)
sns.heatmap(corr, annot=True, mask=mask, cmap="YlGnBu")
plt.show()
```

```python
df.shape
```

DROPING COLUMNS NOT SHOWING MUCH EFFECT ON CO2 EMMISSION

```python
df1 = df.drop(columns=encoded_cols)
#df1= cdf.drop(columns=["Fuel Type_D","Fuel Type_E","Fuel Type_N","Fuel Type_X","Fuel Type_Z","Transmission_A10","Transmission_A4","Transmission_A5","Transmission_A6","Transmission_A7","Transmission_A8","Transmission_A9","Transmission_AM5","Transmission_AM6","Transmission_AM7","Transmission_AM8","Transmission_AM9","Transmission_AS10","Transmission_AS4","Transmission_AS5","Transmission_AS6","Transmission_AS7","Transmission_AS8","Transmission_AS9","Transmission_AV","Transmission_AV10","Transmission_AV6","Transmission_AV7","Transmission_AV8","Transmission_M5","Transmission_M6","Transmission_M7"])
df1
```

REMOVING OUTLAYERS

```python
z = np.abs(stats.zscore(df1))

threshold = 3

#Keep rows with Z-score less than 3
cdf = df1[(z < 3).all(axis=1)]
```

```python
cdf.shape
```

VISUALISATION

```python
plt.scatter(cdf['Fuel Consumption Comb (L/100 km)'], cdf['CO2 Emissions(g/km)'],  color='red')
plt.xlabel("Fuel Consumption Comb")
plt.ylabel("Emission")
plt.show()
```

```python
sns.barplot(x="Engine Size(L)",y="CO2 Emissions(g/km)",data=df1)
matplotlib.rcParams['figure.figsize'] = (40, 10)
```

```python
sns.barplot(x="Cylinders",y="CO2 Emissions(g/km)",data=df1)
matplotlib.rcParams['figure.figsize'] = (20, 10)
```

SCALING

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
col = ['Engine Size(L)',	'Cylinders',	'Fuel Consumption Comb (L/100 km)',	'CO2 Emissions(g/km)']
sc = cdf[col].copy()
scaler.fit(sc)
scaling_col = list(scaler.get_feature_names_out(col))
sc[scaling_col] = scaler.transform(sc)
```

```python
sc
```

```python
X = sc.drop(['CO2 Emissions(g/km)'], axis= 1)
y = sc["CO2 Emissions(g/km)"]
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
print('Training Accuracy: ', lin_reg.score(X_train, y_train))
print('Testing Accuracy: ', lin_reg.score(X_test, y_test))
```

```python
cross_val_score(lin_reg, X, y, cv = 10, scoring = "r2").mean()
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import math
```

```python
pred1 = lin_reg.predict(X_test)
```

```python
r2_scor1=r2_score(y_test, pred1)
print('rmse : {}\nr2 Score : {}'.format(math.sqrt(mean_squared_error(y_test, pred1)),r2_scor1))
```

###Lasso

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso = Lasso()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
lasso_regressor=GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(X_train, y_train)
print(lasso_regressor.best_params_['alpha'])
lassos = Lasso(alpha = lasso_regressor.best_params_['alpha'] ).fit(X_train, y_train)
print('Training Accuracy: ', lassos.score(X_train, y_train))
print('Testing Accuracy: ', lassos.score(X_test, y_test))
```

```python
pred2 = lassos.predict(X_test)
```

```python
r2_scor2=r2_score(y_test, pred2)
print('rmse : {}\nr2 Score : {}'.format(math.sqrt(mean_squared_error(y_test, pred2)),r2_scor2))
```

### Ridge

```python
from sklearn.linear_model import RidgeCV, Ridge
ridge_cv = RidgeCV(alphas = 10**np.linspace(10,-10,100), scoring = "neg_mean_squared_error", normalize = True)
ridge_cv.fit(X_train, y_train)
print('Alpha= ', ridge_cv.alpha_)
ridge_tuned = Ridge(alpha = ridge_cv.alpha_,  normalize = True).fit(X_train,y_train)
print('Training Accuracy: ', ridge_tuned.score(X_train, y_train))
print('Testing Accuracy: ', ridge_tuned.score(X_test, y_test))
```

```python
pred6 = ridge_cv.predict(X_test)
```

```python
r2_scor6=r2_score(y_test, pred6)
print('rmse : {}\nr2 Score : {}'.format(math.sqrt(mean_squared_error(y_test, pred6)),r2_scor6))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state = 42)
rf.fit(X_train, y_train)
print('Training Accuracy: ', rf.score(X_train, y_train))
print('Testing Accuracy: ', rf.score(X_test, y_test))
```

```python
pred3 = rf.predict(X_test)
```

```python
r2_scor3=r2_score(y_test, pred3)
print('rmse : {}\nr2 Score : {}'.format(math.sqrt(mean_squared_error(y_test, pred3)),r2_scor3))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
de = DecisionTreeRegressor(random_state=42)
de.fit(X_train, y_train)
print('Training Accuracy: ', de.score(X_train, y_train))
print('Testing Accuracy: ', de.score(X_test, y_test))
```

```python
from sklearn.tree import plot_tree, export_text
plt.figure(figsize=(80,20))
plot_tree(de, feature_names=X_train.columns, max_depth=2, filled=True);
```

```python
pred4 = de.predict(X_test)
```

```python
r2_scor4=r2_score(y_test, pred4)
print('rmse : {}\nr2 Score : {}'.format(math.sqrt(mean_squared_error(y_test, pred4)),r2_scor4))
```

### KNN

```python
from sklearn.neighbors import KNeighborsRegressor

scores=[]
rmses = []
for i in range(1,10):
  knn = KNeighborsRegressor(n_neighbors=i)
  knn.fit(X_train, y_train)
  scores.append(knn.score(X_test, y_test))
  pre = knn.predict(X_test)
  rmses.append(math.sqrt(mean_squared_error(y_test, pre)))
print(scores)

plt.subplot(1, 2, 1)
plt.plot(list(range(1, len(scores)+1)), scores)
plt.subplot(1, 2, 2)
plt.plot(list(range(1, len(rmses)+1)), rmses)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor

neigh = KNeighborsRegressor(n_neighbors=3)
neigh.fit(X_train, y_train)
print('Training Accuracy: ', neigh.score(X_train, y_train))
print('Testing Accuracy: ', neigh.score(X_test, y_test))
```

```python
pred5 = neigh.predict(X_test)
```

```python
r2_scor5=r2_score(y_test, pred5)
print('rmse : {}\nr2 Score : {}'.format(math.sqrt(mean_squared_error(y_test, pred5)),r2_scor5))
```
